```python
%matplotlib notebook
%precision 3

from IPython.display import set_matplotlib_formats
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import matplotlib

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
plt.rcParams['figure.figsize'] = [9, 5]

import pandas

from datetime import datetime
```

```python
data=pandas.read_csv("78.33.253.139.csv")
data["timestamp"]=pandas.to_datetime(data["UTIME"], unit='s')
data.set_index("timestamp", inplace=True)
data.loc['2020-07-11 00:00':'2020-07-11 06:00']
```

```python
plt.figure()
period = "120s"
rttdata = data.loc['2020-07-10 07:00':'2020-07-11 6:00']["RTT"]
plt.plot(rttdata, label="RTT", linewidth=0.5)


#avgRTT=rttdata.rolling(period)
#avgRTT.mean().plot(label=period +" average RTT", linewidth=1)

plt.ylabel("ms");
plt.xlabel("Time")

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_minor_formatter(mdates.DateFormatter("%H"))
ax.xaxis.set_minor_locator(mdates.HourLocator())
ax.xaxis.grid(True, which='minor')
ax.tick_params(axis='x', which='major', length=15)
ax.tick_params(axis='x', which='minor', length=4)
plt.xticks(rotation=90)
plt.legend()
```

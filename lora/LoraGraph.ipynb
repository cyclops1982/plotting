```python
%matplotlib notebook
%precision 3

from IPython.display import set_matplotlib_formats
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import matplotlib

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.ticker as ticker

plt.rcParams['figure.figsize'] = [9, 4]
plt.rcParams['timezone'] = 'Europe/London'

import pandas

from datetime import datetime
```

```python
data=pandas.read_csv("csvdata.csv",skipinitialspace=True)
data["datetime"]=pandas.to_datetime(data["Unixtime"], unit='s')
data.set_index("datetime", inplace=True)
data
data = data['2019-01-01':]

```

```python
plt.figure()


plt.plot(data["CalcVoltage"], label = "Voltage")
ax = plt.gca()

ax.xaxis.grid(True, which='both')
ax.tick_params(axis='x',which='major', length=35)
ax.tick_params(axis='x',which='minor', rotation=90)

ax.yaxis.set_major_locator(ticker.IndexLocator(1,0));
ax.yaxis.grid(True, which='major')
ax.set_ylim([3, 4])
ax.axhline(0, color='black', lw=1)


ax2 = ax.twinx();
ax2.plot(data["SecondsToFix"], label = "SecondsToFix", linewidth=1, color='blue')
ax2.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
ax2.xaxis.set_major_locator(mdates.DayLocator())
ax2.xaxis.set_minor_formatter(mdates.DateFormatter("%H:%M"))
ax2.xaxis.set_minor_locator(mdates.HourLocator())
ax2.xaxis.grid(True, which='both')

plt.legend()
```
